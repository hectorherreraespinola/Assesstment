### 2) Data Cleaning
Since I did not find specific information about the meaning of the booleans value for the several chronic conditions, I will assume that the value 2 means that the patient has the condition and the value 1 means that the patient does not have the condition. In order to provide a justification for this assumption, I am checking the distribution of the values in the columns that represent the chronic conditions. Since we are analysing data of people with age more than 65, it is reasonable to assume that the value 2 means that the patient has the condition and the value 1 means that the patient does not have the condition.


```python
chronic_conditions_columns = ['SP_ALZHDMTA', 'SP_CHF', 'SP_CHRNKIDN', 'SP_CNCR', 'SP_COPD', 'SP_DEPRESSN', 'SP_DIABETES','SP_ISCHMCHT', 'SP_OSTEOPRS', 'SP_RA_OA', 'SP_STRKETIA']

# Calculate the value counts for each column and plot the distributions
for column in chronic_conditions_columns:
    value_counts = df_member_benefit[column].value_counts()
    value_counts.plot(kind='bar')
    plt.title(f'Distribution of values in the column {column}')
    plt.xlabel('Value')
    plt.ylabel('Count')
    plt.show()
```

```python
#Converting columns for chronic conditions to boolean
df_member_benefit[chronic_conditions_columns] = df_member_benefit[chronic_conditions_columns].replace({2: True, 1: False})
df_member_benefit[chronic_conditions_columns] = df_member_benefit[chronic_conditions_columns].infer_objects(copy=False)# to ensure that the columns are converted to boolean
```

```python
# Function to concatenate diagnoses
def concatenate_diagnoses(row):
    diagnoses = [col for col in chronic_conditions_columns if row[col]]
    if len(diagnoses) >= 3:
        return 'Multiple'
    return ', '.join(diagnoses) if diagnoses else 'None'

# Apply the function to create the new column
df_member_benefit['Chronic_Conditions'] = df_member_benefit.apply(concatenate_diagnoses, axis=1)
```

```python
df_merged = pd.merge(df_member_benefit, df_oupatient_claims, on='DESYNPUF_ID', how='inner')
```

### What is the distribution of the races?

```python
race_labels = {
    0: 'Unknown',
    1: 'White',
    2: 'Black',
    3: 'Other',
    4: 'Asian',
    5: 'Hispanic',
    6: 'North American Native'
}
race_distribution = df_merged['BENE_RACE_CD'].value_counts()


labels = [race_labels.get(code, 'Unknown') for code in race_distribution.index]
sizes = race_distribution.values

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Distribution of Race')
plt.axis('equal')
plt.show()
```

### What is the most common chronic illness combination?


```python
chronic_conditions_counts = df_merged["Chronic_Conditions"].value_counts()
chronic_conditions_counts = chronic_conditions_counts[chronic_conditions_counts.index != 'Multiple']
chronic_conditions_counts = chronic_conditions_counts.sort_values(ascending=False)


highest_condition = chronic_conditions_counts.idxmax()
highest_count = chronic_conditions_counts.max()


print(f"The chronic condition with the highest count is: {highest_condition}")
print(f"The highest count is: {highest_count}")
```

### Which chronic illness combination has the total highest cost?


```python
total_costs_by_condition = df_merged[df_merged['Chronic_Conditions'] != 'Multiple'].groupby('Chronic_Conditions')['TOTAL_COST'].sum()


highest_cost_condition = total_costs_by_condition.idxmax()
highest_cost_value = total_costs_by_condition.max()


print(f"The chronic illness combination with the highest total cost is: {highest_cost_condition}")
print(f"The total cost for this combination is: ${highest_cost_value}")
```

### What is the average cost per member for the chronic illness combination with the highest total cost?

```python

costs_and_counts = df_merged[df_merged['Chronic_Conditions'] != 'Multiple'].groupby('Chronic_Conditions').agg(
    total_cost=('TOTAL_COST', 'sum'),
    member_count=('DESYNPUF_ID', 'nunique')
)

costs_and_counts['avg_cost_per_member'] = costs_and_counts['total_cost'] / costs_and_counts['member_count']


highest_avg_cost_condition = costs_and_counts['avg_cost_per_member'].idxmax()
highest_avg_cost_value = costs_and_counts['avg_cost_per_member'].max()


print(f"The chronic illness combination with the highest cost per member is: {highest_avg_cost_condition}")
print(f"The average cost per member for this combination is: ${highest_avg_cost_value:.2f}")
```

### For each chronic illness combination, represent the distribution of costs per provider.

```python
costs_by_condition_provider = df_merged.groupby(['Chronic_Conditions', 'AT_PHYSN_NPI'])['TOTAL_COST'].sum().reset_index()

costs_by_condition_provider = costs_by_condition_provider[costs_by_condition_provider['Chronic_Conditions'] != 'Multiple']

plt.figure(figsize=(12, 8))
sns.violinplot(x='Chronic_Conditions', y='TOTAL_COST', data=costs_by_condition_provider)
plt.xticks(rotation=90)
plt.title('Distribution of Costs per Provider for Each Chronic Illness Combination')
plt.xlabel('Chronic Illness Combination')
plt.ylabel('Total Cost')
plt.show()
```

```python
costs_and_counts_by_provider = df_merged.groupby(['AT_PHYSN_NPI', 'Chronic_Conditions']).agg(
    total_cost=('TOTAL_COST', 'sum'),
    member_count=('DESYNPUF_ID', 'nunique')
).reset_index()

costs_and_counts_by_provider = costs_and_counts_by_provider[costs_and_counts_by_provider['member_count'] > 1]

costs_and_counts_by_provider['avg_cost_per_member'] = costs_and_counts_by_provider['total_cost'] / costs_and_counts_by_provider['member_count']

avg_cost_per_provider = costs_and_counts_by_provider.groupby('AT_PHYSN_NPI')['avg_cost_per_member'].mean().reset_index()

threshold = avg_cost_per_provider['avg_cost_per_member'].mean() + avg_cost_per_provider['avg_cost_per_member'].std()

expensive_providers = avg_cost_per_provider[avg_cost_per_provider['avg_cost_per_member'] > threshold]

print(expensive_providers)
```

### Which providers are consistently expensive across chronic illnesses they treat?
